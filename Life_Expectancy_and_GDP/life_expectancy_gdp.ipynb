# Life Expectancy and GDP

```python
#libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load Data

```python
df =pd.read_csv("./data/life_expectancy_GDP.csv")
df.head()
```

```python
df.rename(columns = {"Life expectancy at birth (years)" : "Life_Expectancy"}, inplace=True)
df.head()
```

```python
df.info()
```

```python
df.Country.unique()
```

```python
fig, ax = plt.subplots()
sns.lineplot(x= df.Life_Expectancy, y= df.GDP, hue= df.Country)
ax.set_yscale('log')
```

```python
fig, ax = plt.subplots()
sns.barplot(y= df.Country, x= df.Life_Expectancy)
```

```python
plt.figure(figsize=(8,6))
sns.lineplot(y= df.GDP, x= df.Year, hue= df.Country)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), ncol=1)
```

```python
plt.figure(figsize=(8,6))
sns.lineplot(y= df.GDP, x= df.Year, hue= df.Country)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), ncol=1)
plt.yscale('log')
```

this is because not all countries have the same life expectancy

```python
fig, ax1 = plt.subplots(figsize=(12,8))
sns.boxplot(y= df.Country, x= df.GDP)
plt.xticks(rotation=45)
ax1.set_yscale('log')
```

```python
fig, axes = plt.subplots(1, 2, sharey=True, figsize=(15, 5))
axes[0] = sns.swarmplot(ax=axes[0], x=df.GDP, y=df.Country)
axes[0].set_xlabel("GDP in Trillions of U.S. Dollars")
axes[1] = sns.swarmplot(ax=axes[1], x=df.Life_Expectancy, y=df.Country)
axes[1].set_xlabel("Life expectancy at birth (years)");
```
